# LEVEL 1
 # TASK:1  Top Cuisines


```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset
df = pd.read_csv(r"C:\Users\varsh\OneDrive\Documents\Dataset cognify.csv")

# Step 2: Explore the dataset
print("First 5 rows of the dataset:")
print(df.head())
print("\nAvailable columns in dataset:")
print(df.columns)

# Step 3: Check for 'Cuisines' column
if 'Cuisines' not in df.columns:
    raise Exception("Column 'Cuisines' not found in dataset. Please check the column name.")

# Step 4: Count cuisine occurrences
cuisine_counts = df['Cuisines'].value_counts()
top_3 = cuisine_counts.head(3)
print("\nTop 3 most common cuisines:")
print(top_3)

# Step 5: Calculate percentages
total_restaurants = len(df)
top_3_percent = (top_3 / total_restaurants) * 100
print("\nPercentage of total restaurants for Top 3 cuisines:")
print(top_3_percent.round(2))

# Step 6a: Bar chart
plt.figure(figsize=(8, 5))
top_3.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Top 3 Most Common Cuisines (Bar Chart)")
plt.xlabel("Cuisine Type")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Step 6b: Pie chart
plt.figure(figsize=(6, 6))
top_3.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99'], explode=[0.05]*3)
plt.title("Top 3 Most Common Cuisines (Pie Chart)")
plt.ylabel("")  # Remove y-label
plt.tight_layout()
plt.show()
```

# TASK: 2 City Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv(r"C:\Users\varsh\OneDrive\Documents\Dataset cognify.csv")

# Replace with your actual file name

# Optional: clean column names to remove leading/trailing spaces
df.columns = df.columns.str.strip()

#  Identify the city with the highest number of restaurants ---
city_counts = df['City'].value_counts()
print("\nCity with the most restaurants:")
print(city_counts.head(1))

# Calculate the average rating for restaurants in each city ---
city_avg_rating = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False)
print("\nAverage rating by city:")
print(city_avg_rating)

# Determine the city with the highest average rating ---
print("\nCity with highest average rating:")
print(city_avg_rating.head(1))

# Plot 1: Top 10 cities by number of restaurants ---
plt.figure(figsize=(10, 6))
city_counts.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Cities by Number of Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot 2: Top 10 cities by average restaurant rating ---
plt.figure(figsize=(10, 6))
city_avg_rating.head(10).plot(kind='bar', color='lightgreen')
plt.title('Top 10 Cities by Average Restaurant Rating')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# TASK: 4 Online Delivery

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv(r"C:\Users\varsh\OneDrive\Documents\Dataset cognify.csv")

# Clean column names
df.columns = df.columns.str.strip()

# Percentage of restaurants offering online delivery ---
online_delivery_counts = df['Has Online delivery'].value_counts()
total = online_delivery_counts.sum()
online_delivery_percentage = (online_delivery_counts.get('Yes', 0) / total) * 100

print(f"Percentage of restaurants offering online delivery: {online_delivery_percentage:.2f}%")

# Average ratings by delivery availability ---
avg_rating = df.groupby('Has Online delivery')['Aggregate rating'].mean().reset_index()

# Bar Plot: Average Ratings with and without Online Delivery
plt.figure(figsize=(6,4))
sns.barplot(data=avg_rating, x='Has Online delivery', y='Aggregate rating', palette='Set2')
plt.title('Average Ratings: Online Delivery vs No Delivery')
plt.ylabel('Average Aggregate Rating')
plt.xlabel('Online Delivery Available')
plt.ylim(0, 5)
plt.tight_layout()
plt.show()

# Histogram: Distribution of ratings based on delivery option
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='Aggregate rating', hue='Has Online delivery', multiple='stack', bins=20, palette='Set2')
plt.title('Rating Distribution by Online Delivery Availability')
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Restaurants')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv(r"C:\Users\varsh\OneDrive\Documents\Dataset cognify.csv")
# Replace with your correct path

# Clean column names (recommended)
df.columns = df.columns.str.strip()

# Use correct column names
if 'Has Online delivery' not in df.columns or 'Aggregate rating' not in df.columns:
    raise Exception("Required columns not found. Check column names.")

# Filter restaurants that offer online delivery
online_delivery = df[df['Has Online delivery'] == 'Yes']
no_online_delivery = df[df['Has Online delivery'] == 'No']

# Calculate percentage
total = len(df)
online_percent = (len(online_delivery) / total) * 100
print(f"\n Percentage of restaurants offering online delivery: {online_percent:.2f}%")

# Compare average ratings
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')
avg_online = online_delivery['Aggregate rating'].mean()
avg_offline = no_online_delivery['Aggregate rating'].mean()

print(f" Average rating (Online delivery): {avg_online:.2f}")
print(f" Average rating (No online delivery): {avg_offline:.2f}")

# Visualize the comparison
plt.bar(['Online Delivery', 'No Online Delivery'], [avg_online, avg_offline], color=['green', 'red'])
plt.title("Average Ratings: Online vs No Online Delivery")
plt.ylabel("Average Rating")
plt.ylim(0, 5)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# TASK 2: Cuisine Combination

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv(r"C:\Users\varsh\OneDrive\Documents\Dataset cognify.csv")
# Replace with your correct path

# Clean data: Drop missing cuisines and convert to lowercase
df['Cuisines'] = df['Cuisines'].dropna().str.lower().str.strip()

# 1. Most common cuisine combinations
top_combinations = df['Cuisines'].value_counts().head(10)
print("Top 10 most common cuisine combinations:\n")
print(top_combinations)

# Plot top 10 cuisine combinations
plt.figure(figsize=(10, 6))
sns.barplot(y=top_combinations.index, x=top_combinations.values, palette='viridis')
plt.title("Top 10 Most Common Cuisine Combinations")
plt.xlabel("Number of Restaurants")
plt.ylabel("Cuisine Combination")
plt.tight_layout()
plt.show()

# 2. Average rating by cuisine combination (top 10 only)
top_combo_names = top_combinations.index.tolist()
top_combo_ratings = df[df['Cuisines'].isin(top_combo_names)].groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)

print("\nAverage ratings of top cuisine combinations:\n")
print(top_combo_ratings)

# Plot average ratings
plt.figure(figsize=(10, 6))
sns.barplot(x=top_combo_ratings.values, y=top_combo_ratings.index, palette='mako')
plt.title("Average Rating by Top Cuisine Combinations")
plt.xlabel("Average Rating")
plt.ylabel("Cuisine Combination")
plt.tight_layout()
plt.show()
```

# TASK: 3 Geographic Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Install and import geopandas & contextily for advanced maps
# !pip install geopandas contextily

# Load the dataset
df = pd.read_csv(r"C:\Users\varsh\OneDrive\Documents\Dataset cognify.csv")

# Clean and convert latitude and longitude columns
df['Latitude'] = pd.to_numeric(df['Latitude'], errors='coerce')
df['Longitude'] = pd.to_numeric(df['Longitude'], errors='coerce')

# Drop rows with missing or invalid coordinates
geo_df = df.dropna(subset=['Latitude', 'Longitude'])

# 1. Scatter plot of restaurant locations
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Longitude', y='Latitude', data=geo_df, alpha=0.4, edgecolor=None)
plt.title("Restaurant Locations")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.tight_layout()
plt.show()

# 2. Optional: Heatmap using seaborn to show density
plt.figure(figsize=(10, 6))
sns.kdeplot(
    x=geo_df['Longitude'], 
    y=geo_df['Latitude'], 
    cmap="Reds", fill=True, thresh=0.05, levels=100
)
plt.title("Restaurant Density Heatmap")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.tight_layout()
plt.show()
```

# TASK: 4 Restaurant Chians

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv(r"C:\Users\varsh\OneDrive\Documents\Dataset cognify.csv")

# Drop null values from Restaurant Name, Aggregate rating, and Votes
df = df.dropna(subset=['Restaurant Name', 'Aggregate rating', 'Votes'])

# Convert votes and ratings to numeric
df['Votes'] = pd.to_numeric(df['Votes'], errors='coerce')
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')

# 1. Identify restaurant chains (those appearing more than once)
chain_counts = df['Restaurant Name'].value_counts()
restaurant_chains = chain_counts[chain_counts > 1].index.tolist()

# Filter dataset to include only chains
chain_df = df[df['Restaurant Name'].isin(restaurant_chains)]

# Group by restaurant chain and compute average rating and votes
chain_summary = chain_df.groupby('Restaurant Name').agg({
    'Aggregate rating': 'mean',
    'Votes': 'sum',
    'Restaurant Name': 'count'  # to show number of branches
}).rename(columns={'Restaurant Name': 'Branch Count'}).sort_values(by='Votes', ascending=False).head(10)

print("Top 10 Restaurant Chains by Popularity (Votes):\n")
print(chain_summary)

# 2. Plot the top 10 chains by popularity
plt.figure(figsize=(12, 6))
sns.barplot(
    x=chain_summary['Votes'],
    y=chain_summary.index,
    palette='coolwarm'
)
plt.title("Top 10 Popular Restaurant Chains (by Total Votes)")
plt.xlabel("Total Votes")
plt.ylabel("Restaurant Chain")
plt.tight_layout()
plt.show()

# 3. Plot average ratings of those chains
plt.figure(figsize=(12, 6))
sns.barplot(
    x=chain_summary['Aggregate rating'],
    y=chain_summary.index,
    palette='crest'
)
plt.title("Average Ratings of Top 10 Restaurant Chains")
plt.xlabel("Average Rating")
plt.ylabel("Restaurant Chain")
plt.tight_layout()
plt.show()
```

# TASK 1: Restaurent Reviews

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import re

# Load dataset
df = pd.read_csv(r"C:\Users\varsh\OneDrive\Documents\Dataset cognify.csv")

# Convert Rating text to lowercase and remove punctuation
df['Rating text'] = df['Rating text'].astype(str).str.lower()
df['Cleaned_Reviews'] = df['Rating text'].apply(lambda x: re.sub(r'[^a-z\s]', '', x))

# Calculate review length in number of words
df['Review_Length'] = df['Cleaned_Reviews'].apply(lambda x: len(x.split()))
print("Average Review Length:", round(df['Review_Length'].mean(), 2), "words")

# Count most common words
all_words = ' '.join(df['Cleaned_Reviews']).split()
common_words = Counter(all_words).most_common(10)
print("\nTop 10 Common Words in Reviews:")
for word, count in common_words:
    print(f"{word}: {count}")

# Plot Review Length vs Aggregate Rating
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')

plt.figure(figsize=(8, 5))
plt.scatter(df['Review_Length'], df['Aggregate rating'], alpha=0.5)
plt.title("Review Length vs. Aggregate Rating")
plt.xlabel("Review Length (words)")
plt.ylabel("Aggregate Rating")
plt.grid(True)
plt.tight_layout()
plt.show()
```

# TASK 2: Voter Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv(r"C:\Users\varsh\OneDrive\Documents\Dataset cognify.csv")

# Convert Rating text to lowercase and remove punctuation

# Ensure numeric data types
df['Votes'] = pd.to_numeric(df['Votes'], errors='coerce')
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')

# Drop rows with missing values in relevant columns
df_clean = df.dropna(subset=['Votes', 'Aggregate rating'])

# 1. Restaurant with highest number of votes
max_votes_row = df_clean[df_clean['Votes'] == df_clean['Votes'].max()]
print("\nRestaurant with Highest Votes:")
print(max_votes_row[['Restaurant Name', 'Votes', 'Aggregate rating']])

# 2. Restaurant with lowest non-zero votes
min_votes_row = df_clean[df_clean['Votes'] > 0].nsmallest(1, 'Votes')
print("\nRestaurant with Lowest (Non-zero) Votes:")
print(min_votes_row[['Restaurant Name', 'Votes', 'Aggregate rating']])

# 3. Correlation between Votes and Aggregate Rating
correlation = df_clean['Votes'].corr(df_clean['Aggregate rating'])
print(f"\nCorrelation between Votes and Aggregate Rating: {correlation:.2f}")

# 4. Scatter plot
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_clean, x='Votes', y='Aggregate rating', alpha=0.6)
plt.title("Votes vs Aggregate Rating")
plt.xlabel("Number of Votes")
plt.ylabel("Aggregate Rating")
plt.grid(True)
plt.tight_layout()
plt.show()
```

# TASK 3: Price Range vs Online Delivery and Table Booking

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv(r"C:\Users\varsh\OneDrive\Documents\Dataset cognify.csv")


# Ensure relevant columns are clean
df['Price range'] = pd.to_numeric(df['Price range'], errors='coerce')
df['Has Online delivery'] = df['Has Online delivery'].astype(str)
df['Has Table booking'] = df['Has Table booking'].astype(str)

# Group by Price Range for Online Delivery
online_delivery_group = df.groupby('Price range')['Has Online delivery'].value_counts(normalize=True).unstack().fillna(0)

# Group by Price Range for Table Booking
table_booking_group = df.groupby('Price range')['Has Table booking'].value_counts(normalize=True).unstack().fillna(0)

# Plot Online Delivery vs Price Range
plt.figure(figsize=(10, 4))
online_delivery_group[['Yes', 'No']].plot(kind='bar', stacked=True, color=['green', 'red'], ax=plt.gca())
plt.title("Online Delivery Availability by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Proportion")
plt.legend(title='Online Delivery')
plt.tight_layout()
plt.show()

# Plot Table Booking vs Price Range
plt.figure(figsize=(10, 4))
table_booking_group[['Yes', 'No']].plot(kind='bar', stacked=True, color=['blue', 'orange'], ax=plt.gca())
plt.title("Table Booking Availability by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Proportion")
plt.legend(title='Table Booking')
plt.tight_layout()
plt.show()
```
